```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
# Part a
n = int(1 / 0.0025**2 / 4) # large enough n
print("n is: %d" % n)    
Z=np.random.randn(n)

# Part b
ki = 0
Y = [0]*4
for k in [1,8,64,512]:
    Y[ki] = np.sum(np.sign(np.random.randn(n, k))*np.sqrt(1./k), axis=1) # generate Y
    ki+=1
```

```python
sns.set(color_codes=True)
plt.figure()
plt.step(sorted(Y[0]), np.arange(1,n+1)/float(n), label = '1')
plt.step(sorted(Y[1]), np.arange(1,n+1)/float(n), label = '8')
plt.step(sorted(Y[2]), np.arange(1,n+1)/float(n), label = '64')
plt.step(sorted(Y[3]), np.arange(1,n+1)/float(n), label = '512')
plt.step(sorted(Z), np.arange(1,n+1)/float(n), label = 'Gaussian')

plt.xlabel('Observations')
plt.ylabel('Probability')

plt.legend(loc='right')

plt.ylim(-0,1)
plt.xlim(-3,3)
plt.show()
```
