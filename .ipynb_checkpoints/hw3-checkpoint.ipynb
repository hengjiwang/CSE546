## 3.a

```python
N = 30 
X = np.random.rand(N)
Y = 4 * np.sin(np.pi * X) * np.cos(6 * np.pi * X**2) + np.random.randn(N)
```

```python
def kernel_poly(x, z, d):
    return (1+x@z.T) ** d

def kernel_rbf(x, z, gamma):
    return np.exp(-gamma*(x-z)*(x-z))

eta = 0.01

# poly kernel
obj_list = []
lamb_list = []
d_list =[]
lamb = 1
while lamb > 1e-4:
    for d in range(0, 100):
        obj = 0
        for i in range(N):
            X_tr = np.append(X[0:i], X[i+1:N])
            Y_tr = np.append(Y[0:i], Y[i+1:N])
            X_va = X[i]
            Y_va = Y[i]
            K = kernel_poly(X_tr[:,np.newaxis],X_tr[:,np.newaxis], d)
            alpha = np.linalg.solve((K + lamb*np.eye(N-1, N-1)), Y_tr)
            K_va = (1+X_va*X_tr[np.newaxis,:]) ** d
            obj += (K_va@alpha - Y_va).T@(K_va@alpha - Y_va)
        obj_list.append(obj)
        lamb_list.append(lamb)
        d_list.append(d) 
    lamb /= 2    
lamb_poly = lamb_list[obj_list.index(np.min(obj_list))]
d_poly = d_list[obj_list.index(np.min(obj_list))]
print('lambda for poly kernel is :', lamb_poly)
print('d for poly kernel is :', d_poly)
```

```python
f_poly=[]
K_poly = kernel_poly(X[:,np.newaxis],X[:,np.newaxis], d_poly)
alpha_poly = np.linalg.solve((K_poly + lamb_poly*np.eye(N, N)), Y)
for xx in np.arange(min(X),max(X),0.001):
    f_poly.append(np.sum((1+xx*X[np.newaxis,:]) ** d_poly @ alpha_poly))
```

```python
# RBF kernel
obj_list = []
lamb_list = []
d_list =[]

dist = []
for xi in X_tr:
    for xj in X_tr:
        dist.append((xi-xj)**2)
gamma = 1/np.median(dist)

lamb = 1
while lamb > 1e-20:
    obj = 0
    for i in range(N):  
        X_tr = np.append(X[0:i], X[i+1:N])
        Y_tr = np.append(Y[0:i], Y[i+1:N])
        X_va = X[i]
        Y_va = Y[i]
        K = kernel_rbf(X_tr[:,np.newaxis],X_tr[np.newaxis,:], gamma)
        alpha = np.linalg.solve((K + lamb*np.eye(N-1, N-1)), Y_tr)
        K_va = np.exp(-gamma*(X_va-X_tr[np.newaxis,:])**2)
        obj += (K_va@alpha - Y_va).T@(K_va@alpha - Y_va)
    obj_list.append(obj)
    lamb_list.append(lamb) 
    lamb /= 2    
lamb_rbf = lamb_list[obj_list.index(np.min(obj_list))]
print('gamma for rbf kernel is : ', gamma)
print('lambda for rbf kernel is :', lamb_rbf)
```

```python
f_rbf = []
f0 = []
K_rbf = kernel_rbf(X[:,np.newaxis],X[np.newaxis,:], gamma)
alpha_rbf = np.linalg.solve((K_rbf + lamb_rbf*np.eye(N, N)), Y)
for xx in np.arange(min(X),max(X),0.001):
    f_rbf.append(np.sum(alpha_rbf * np.exp(-gamma*(xx-X)**2)))
    f0.append(4*np.sin(np.pi*xx)*np.cos(6*np.pi*xx**2))
```

```python
import numpy as np
from mnist import MNIST
import math
import matplotlib.pyplot as plt
import random

mndata = MNIST('./data/')
X_train, labels_train = map(np.array, mndata.load_training())
X_test, labels_test = map(np.array, mndata.load_testing())
X_train = X_train/255.0
X_test = X_test/255.0
```

```python
def div_clusters(X, centers):
    clusters = [[] for j in range(len(centers))]
    obj = 0
    for x in X:
        dist = []
        for i in range(len(centers)):
            dist.append(np.linalg.norm(x-centers[i]))
        obj += np.min(dist)**2
        closest = dist.index(np.min(dist))
        #try:
        clusters[closest].append(x)
        #except KeyError:
        #   clusters[closest] = [x]
    return clusters, obj

def calc_centers(clusters):
    centers = []
    #keys = sorted(clusters.keys())
    for i in range(len(clusters)):
        centers.append(np.mean(clusters[i], axis = 0))
    return centers

def is_converge(c1, c2):
    flag = []
    for i in range(len(c1)):
        if np.sum(c1[i] != c2[i]) > 0:
            return False
    return True
```

```python
centers_cub = []
objss = []
for K in [5, 10, 20]:
    objs = []
    old_centers = random.sample(list(X_train), K)
    centers = random.sample(list(X_train), K)
    while not is_converge(old_centers, centers):
        old_centers = centers
        clusters, obj = div_clusters(X_train, centers)
        centers = calc_centers(clusters)
        objs.append(obj)
        #print(obj)
    centers_cub.append(centers)
    objss.append(objs)
    print('Done:', K)
```

```python
def random_center(X, prob):
    picker = np.random.uniform(0,1)
    cumsum = 0
    for j in range(len(prob)):
        cumsum += prob[j]
        if picker < cumsum:
            return X[j]
```

```python
def init_centers(X, K):
    centers = []
    centers.append(random.sample(list(X), 1))
    for j in range(1,K):
        ii = 0
        shortest_dist2 = np.zeros(len(X))
        for x in X:
            dist = []
            for i in range(len(centers)):
                dist.append(np.linalg.norm(x-centers[i]))
            closest = dist.index(np.min(dist))
            shortest_dist2[ii] = np.linalg.norm(x-centers[closest]) ** 2
            ii += 1
        total_dist2 = np.sum(shortest_dist2)
        prob = np.divide(shortest_dist2, total_dist2)
        centers.append(random_center(X, prob))
    return centers
```

```python
centers_cub = []
objss = []
for K in [5, 10, 20]:
    objs = []
    old_centers = random.sample(list(X_train), K)
    centers = init_centers(X_train, K)
    
    while not is_converge(old_centers, centers):
        old_centers = centers
        clusters, obj = div_clusters(X_train, centers)
        centers = calc_centers(clusters)
        objs.append(obj)
        print(obj)
    centers_cub.append(centers)
    objss.append(objs)
    print('Done:', K)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.sparse.linalg as SLA
```

```python
file_train = open('./jester/train.txt', 'r')
data_train_str = file_train.readlines()
for j in range(len(data_train_str)):
    data_train_str[j] = data_train_str[j].strip('\n').split(',')
file_train.close()       

file_test = open('./jester/test.txt', 'r')
data_test_str = file_test.readlines()
for j in range(len(data_test_str)):
    data_test_str[j] = data_test_str[j].strip('\n').split(',')
file_test.close()   

data_train = []
for line in data_train_str:
    data_train.append(list(map(eval, line)))
    
data_test = []
for line in data_test_str:
    data_test.append(list(map(eval, line)))
    
data_train = np.array(data_train)
data_test = np.array(data_test)
```

## 5.a

```python
nUser = 24983
nJoke = 100

avrgs = np.zeros(nJoke)
users = np.zeros(nJoke)
for j in range(len(data_train)):
    avrgs[int(data_train[j,1])-1] += data_train[j,2]
    users[int(data_train[j,1])-1] += 1
avrgs = np.divide(avrgs, users)
```

## 5.b

```python
R_train = np.zeros((nUser, nJoke))
for j in range(len(data_train)):
    R_train[int(data_train[j,0])-1, int(data_train[j,1])-1] = data_train[j,2]   
    
R_test = np.zeros((nUser, nJoke))
for j in range(len(data_test)):
    R_test[int(data_test[j,0])-1, int(data_test[j,1])-1] = data_test[j,2] 
```

```python
tr_sqr_error = []
tr_abs_error = []
te_sqr_error = []
te_abs_error = []
d_list = [1,2,5,10,20,50]
for d in d_list:
    u, s, vt = SLA.svds(R_train, d)
    z = u@np.diag(s)@vt
    #z = u@vt
    tr_err = 0
    te_err = 0
    for i in range(nUser):
        for j in range(nJoke):
            if R_train[i][j] == 0:
                tr_err += (z[i][j] - R_train[i][j])**2
            if R_test[i][j] == 0:
                te_err += (z[i][j] - R_test[i][j])**2
    te_err = 0
    for j in range(len(R_test)):
        tr_err += np.sum(np.abs(z[j]-R_train[j]))/np.count_nonzero(R_train[j])
        te_err += np.sum(np.abs(z[j]-R_test[j]))/np.count_nonzero(R_test[j])
    tr_abs_error.append(tr_err/nUser)
    te_abs_error.append(te_err/nUser)

#tr_sqr_error = np.divide(tr_sqr_error, nUser*nJoke)
#te_sqr_error = np.divide(te_sqr_error, nUser*nJoke)
tr_sqr_error = np.divide(tr_sqr_error, len(data_train))
te_sqr_error = np.divide(te_sqr_error, len(data_test))
```

```python
plt.figure()
plt.plot(d_list, tr_sqr_error, 'bo-', label = 'train square')
plt.plot(d_list, tr_abs_error,'b^-', label = 'train abs')
plt.plot(d_list, te_sqr_error, 'ro-', label = 'test square')
plt.plot(d_list, te_abs_error,'r^-', label = 'test abs')
plt.xlabel('d')
plt.ylabel('error metrics')
plt.legend(loc='upper left')
plt.show()
```
