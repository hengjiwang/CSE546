## 3.a

```python
N = 30 
X = np.random.rand(N)
Y = 4 * np.sin(np.pi * X) * np.cos(6 * np.pi * X**2) + np.random.randn(N)
```

```python
def kernel_poly(x, z, d):
    return (1+x@z.T) ** d

def kernel_rbf(x, z, gamma):
    return np.exp(-gamma*(x-z)*(x-z))

eta = 0.01

# poly kernel
obj_list = []
lamb_list = []
d_list =[]
lamb = 1
while lamb > 1e-4:
    for d in range(0, 100):
        obj = 0
        for i in range(N):
            X_tr = np.append(X[0:i], X[i+1:N])
            Y_tr = np.append(Y[0:i], Y[i+1:N])
            X_va = X[i]
            Y_va = Y[i]
            K = kernel_poly(X_tr[:,np.newaxis],X_tr[:,np.newaxis], d)
            alpha = np.linalg.solve((K + lamb*np.eye(N-1, N-1)), Y_tr)
            K_va = (1+X_va*X_tr[np.newaxis,:]) ** d
            obj += (K_va@alpha - Y_va).T@(K_va@alpha - Y_va)
        obj_list.append(obj)
        lamb_list.append(lamb)
        d_list.append(d) 
    lamb /= 2    
lamb_poly = lamb_list[obj_list.index(np.min(obj_list))]
d_poly = d_list[obj_list.index(np.min(obj_list))]
print('lambda for poly kernel is :', lamb_poly)
print('d for poly kernel is :', d_poly)
```

```python
f_poly=[]
K_poly = kernel_poly(X[:,np.newaxis],X[:,np.newaxis], d_poly)
alpha_poly = np.linalg.solve((K_poly + lamb_poly*np.eye(N, N)), Y)
for xx in np.arange(min(X),max(X),0.001):
    f_poly.append(np.sum((1+xx*X[np.newaxis,:]) ** d_poly @ alpha_poly))
```

```python
# RBF kernel
obj_list = []
lamb_list = []
d_list =[]

dist = []
for xi in X_tr:
    for xj in X_tr:
        dist.append((xi-xj)**2)
gamma = 1/np.median(dist)

lamb = 1
while lamb > 1e-20:
    obj = 0
    for i in range(N):  
        X_tr = np.append(X[0:i], X[i+1:N])
        Y_tr = np.append(Y[0:i], Y[i+1:N])
        X_va = X[i]
        Y_va = Y[i]
        K = kernel_rbf(X_tr[:,np.newaxis],X_tr[np.newaxis,:], gamma)
        alpha = np.linalg.solve((K + lamb*np.eye(N-1, N-1)), Y_tr)
        K_va = np.exp(-gamma*(X_va-X_tr[np.newaxis,:])**2)
        obj += (K_va@alpha - Y_va).T@(K_va@alpha - Y_va)
    obj_list.append(obj)
    lamb_list.append(lamb) 
    lamb /= 2    
lamb_rbf = lamb_list[obj_list.index(np.min(obj_list))]
print('lambda for rbf kernel is :', lamb_rbf)
```

```python
f_rbf = []
f0 = []
K_rbf = kernel_rbf(X[:,np.newaxis],X[np.newaxis,:], gamma)
alpha_rbf = np.linalg.solve((K_rbf + lamb_rbf*np.eye(N, N)), Y)
for xx in np.arange(min(X),max(X),0.001):
    f_rbf.append(np.sum(alpha_rbf * np.exp(-gamma*(xx-X)**2)))
    f0.append(4*np.sin(np.pi*xx)*np.cos(6*np.pi*xx**2))
```

## 3.b

```python
aa = []
for j in range(len(f0)):
    aa.append(min(X)+0.001*j)
plt.figure
plt.plot(aa,f0, label = 'true f(x)')
plt.plot(aa,f_poly, label = 'poly f(x)')
plt.plot(X,Y,'ro', label ='original data')
plt.legend(loc='lower left')
plt.ylim(-6,6)
plt.show()

plt.figure
plt.plot(aa,f0, label = 'true f(x)')
plt.plot(aa,f_rbf,'g-', label = 'rbf f(x)')
plt.plot(X,Y,'ro', label ='original data')
plt.legend(loc='lower left')
plt.ylim(-6,6)
plt.show()
```

```python
import numpy as np
from mnist import MNIST
import math
import matplotlib.pyplot as plt
import random

mndata = MNIST('./data/')
X_train, labels_train = map(np.array, mndata.load_training())
X_test, labels_test = map(np.array, mndata.load_testing())
X_train = X_train/255.0
X_test = X_test/255.0
```

```python
def div_clusters(X, centers):
    clusters = [[] for j in range(len(centers))]
    obj = 0
    for x in X:
        dist = []
        for i in range(len(centers)):
            dist.append(np.linalg.norm(x-centers[i]))
        obj += np.min(dist)
        closest = dist.index(np.min(dist))
        #try:
        clusters[closest].append(x)
        #except KeyError:
        #   clusters[closest] = [x]
    return clusters, obj

def calc_centers(clusters):
    centers = []
    #keys = sorted(clusters.keys())
    for i in range(len(clusters)):
        centers.append(np.mean(clusters[i], axis = 0))
    return centers

def is_converge(c1, c2):
    flag = []
    for i in range(len(c1)):
        if np.sum(c1[i] != c2[i]) > 0:
            return False
    return True
```

```python
centers_cub = []
objss = []
for K in [5, 10, 20]:
    objs = []
    old_centers = random.sample(list(X_train), K)
    centers = random.sample(list(X_train), K)
    while not is_converge(old_centers, centers):
        old_centers = centers
        clusters, obj = div_clusters(X_train, centers)
        centers = calc_centers(clusters)
        objs.append(obj)
        #print(obj)
    centers_cub.append(centers)
    objss.append(objs)
    print('Done:', K)
```

### Plot K = 5

```python
xaxis = []
for j in range(len(objss[0])):
    xaxis.append(j)
#plt.figure(figsize=(10, 6))
plt.plot(xaxis, objss[0], 'bo-')
plt.xlabel('iteration steps')
plt.xlabel('objective function')
plt.title('objective function for k=5')
plt.show()
```

```python
N = 9*160
ax = [0 for j in range(5)]
for j in range(len(ax)):
    fig, ax[j] = plt.subplots()
    #ax[j].imshow(X_train[j].reshape(28,28), cmap=plt.cm.binary, interpolation='none', extent=[0, N, 0, N], zorder=2)
    ax[j].imshow(centers_cub[0][j].reshape(28,28), cmap=plt.cm.binary, interpolation='none', extent=[0, N, 0, N], zorder=6)
    ax[j].axis('off')
plt.show()
```
